```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['font.family'] = 'serif'
plt.rcParams['text.latex.preamble'] = r'\usepackage{amsmath,amssymb}'
```

```python
def generate_points_on_sphere(dim, points):
    x = np.random.normal(0.0, 1.0, size=(points, dim))
    x /= np.sqrt(np.sum(x*x, axis=1, keepdims=True))
    #x[i,j] = jth component of the ith sample

    dots = x @ x.T
    d = 2.0 - 2.0 * dots
    d = np.maximum(d, 0.0) 
    d = np.sqrt(d)
    #d[i,j] = distance between ith and jth sample

    return x, d

#Points on sphere in 3D
X1, d1 = generate_points_on_sphere(3,100)

#Points on sphere in 100D
X2, d2 = generate_points_on_sphere(100,100)

np.fill_diagonal(d1, np.nan)
np.fill_diagonal(d2, np.nan)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(6, 5))
fig.set_dpi(200)
axes[0].hist(d1.flatten(), bins=50, density=False)
axes[0].set_ylabel("Number of Points $(d=3)$")
axes[1].hist(d2.flatten(), bins=50, density=False)
axes[1].set_ylabel("Number of Points $(d=100)$")
axes[1].set_xlabel("Pairwise Distance")
```
